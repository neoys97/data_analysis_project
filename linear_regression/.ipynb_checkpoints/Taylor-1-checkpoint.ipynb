```python
#A) Import data
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv('airbnb data.csv')
subset_df = df.loc[:, ['accommodates', 'bathrooms','neighbourhood','num_beds','price','property_type']]
subset_df
```

```python
#B) Remove duplicates
noDuplicates_df = subset_df.drop_duplicates()
noDuplicates_df
```

```python
#B) Drop records with missing values
noNull_df = noDuplicates_df.dropna()
noNull_df
```

```python
#C) Split into training and testing data
from sklearn.model_selection import train_test_split
x = noNull_df.loc[:, ['accommodates', 'bathrooms', 'neighbourhood', 'num_beds', 'price', 'property_type']]
y = noNull_df.loc[:, ["price"]]
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 10)
x_train
```

```python
loc_dict = {}
counter = 0;

for i in range(0, 500):
    if i in noNull_df['neighbourhood']:
        if noNull_df['neighbourhood'][i] in loc_dict:
            noNull_df['neighbourhood'][i] = loc_dict[noNull_df['neighbourhood'][i]]
        else:
            loc_dict[noNull_df['neighbourhood'][i]] = counter
            noNull_df['neighbourhood'][i] = counter
            counter = counter + 1
        
print(loc_dict)

# noNull_df['neighbourhood'] = pd.to_numeric(noNull_df['neighbourhood'], errors='coerce').fillna(0, downcast='infer')
    
noNull_df['neighbourhood']
```
